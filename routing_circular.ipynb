```python
import random
from helper_circular import *
```

```python
# Define cylinder parameters
cylinder_params = {
    'radius_inner': 28,
    'radius_outer': 30,
    'num_cylinders': 8,
    'cylinder_circle_radius': 20,
    'cylinder_height': 12,
    'cylinder_radius': 3.8,
    'tube_radius': 1.5,
    'resolution': 1,
}

top_list, bottom_list = generate_cylinder_points(cylinder_params)

# Set random seed for reproducibility
random.seed(42)
# Generate points
random.shuffle(bottom_list)

connections = list(zip(top_list, bottom_list))
```

```python
# Set parameters
min_separation = 5  # Minimum separation distance between paths

# Find paths
paths, failed_paths = find_all_paths(connections, cylinder_params, min_separation)

# Display results
print(f"\nResults:")
print(f"  {len(paths)} valid paths generated")
print(f"  {len(failed_paths)} failed paths")
```

```python
# Visualize paths (if applicable)
if paths:
    visualize_paths(paths, connections[:len(paths)], cylinder_params)
```

```python
# Export to OpenSCAD
export_openscad(paths, "cylindrical_paths_normal.scad")
```
